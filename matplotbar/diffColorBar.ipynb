```python
"""
Title - Write a Python programming to display a bar chart of the popularity of programming
        Languages. Use different color for each bar.
        Sample data:
        Programming languages: Java, Python, PHP, JavaScript, C#, C++
        Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
Author name - Aditya Kumar
Creation time - 15 March 2021 
Modified time - 15 March 2021

"""
import matplotlib.pyplot as plt

progLang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++', 'Android']
Popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7, 18.9]
myColor = ['r','g','b','k','y','m','c']

plt.style.use('ggplot')
plt.tight_layout()

plt.title('Popular Programming Languages')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')

plt.bar(progLang, Popularity, width = .4 ,color = myColor )

```
