```python

"""
Title - Write a Python programming to display a bar chart of the popularity of programming
        Languages. Increase bottom margin.
        Sample data:
        Programming languages: Java, Python, PHP, JavaScript, C#, C++
        Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
Author name - Aditya Kumar
Creation time - ‎‎15 ‎March ‎2021 ‏‎
Modified time - ‎‎‎15 ‎March ‎2021‎

"""

import matplotlib.pyplot as plt

progLang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
Popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(progLang)]

plt.style.use('ggplot')
plt.tight_layout()
plt.xticks(x_pos, progLang, rotation=90)

plt.title('Popular Programming Languages')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')

plt.bar(x_pos, Popularity, width = .4 , color = 'g')
plt.subplots_adjust(bottom = .5, top = 1)
```
