```python
"""
Title - Write a program to draw a scatter plot for random 500 x and y coordinates and style it.
Author name - Aditya Kumar
Creation time - 17 March 2021 
Modified time - 17 March 2021

"""

import numpy as np
import pandas as pd
import plotly.express as px

n =500

arr1 = np.random.randn(n,3)
df = pd.DataFrame(arr1 , columns = ['A', 'B','C'])

fig = px.scatter(x = df['A'], y = df['B'] , color = df['C'])
fig.show()
```
