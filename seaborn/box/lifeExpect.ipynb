```python
"""
Title - Write a program to draw box plot of life expectancy by continent for a data set given in a url.
        https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv
Author name - Aditya Kumar
Creation time - ‎‎16 ‎March ‎2021 ‏‎
Modified time - ‎‎‎16 ‎March ‎2021‎

"""
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

urll = 'https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv'

df = pd.read_csv(urll)
df.head()
```

```python
sns.boxplot(x = 'lifeExp', y = 'continent' , data = df)
plt.show()
```
