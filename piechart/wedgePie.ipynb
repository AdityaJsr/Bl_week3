```python
"""
Title - Write a Python programming to create a pie chart with a title of the popularity of programming
        Languages.Make multiple wedges of the pie.
        Sample data:
        Programming languages: Java, Python, PHP, JavaScript, C#, C++
        Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

Author name - Aditya Kumar
Creation time - 15 March 2021 
Modified time - 16 March 2021

"""

import matplotlib.pyplot as plt

# Data
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
explode = (0.2, 0.1, 0.1, 0.1, 0.1, .1) 

# Plot
plt.pie(popuratity, explode = explode, labels=languages, autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')

plt.show()
```
