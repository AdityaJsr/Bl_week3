```python

"""
Title - Write a Python program to plot two or more lines on same plot with suitable legends of each
        line.
Author name - Aditya Kumar
Creation time - 13 March 2021 
Modified time - 13 March 2021

"""

import matplotlib.pyplot as plt
# Data for the two lines.
x1 = [10,20,30]
x2 = [10,20,30]
y1 = [20,40,10]
y2 = [40,10,30]

# Plotting both the points with label.

plt.plot(x1, y1, label = "line 1")
plt.plot(x2, y2, label = "line 2")

# Labeling axis and giving title.

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Two or more lines on same plot with suitable legends ')

# Creating legends and showing the output.

plt.legend()
plt.show()



```
