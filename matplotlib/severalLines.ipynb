```python
"""
Title - Write a Python program to plot several lines with different format styles in one command
using arrays.
Author name - Aditya Kumar
Creation time - ‎‎13 ‎March ‎2021 ‏‎
Modified time - ‎‎‎14 ‎March ‎2021‎

"""


import numpy as np
import matplotlib.pyplot as plt

# Sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# green dashes, blue squares and red triangles
plt.plot(t, t, 'go', t, t**2, 'bs', t, t**3, 'r^')
plt.show()
```
