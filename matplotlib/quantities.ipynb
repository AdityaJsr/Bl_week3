```python
"""
Title - Write a Python program to plot quantities which have an x and y position.
Author name - Aditya Kumar
Creation time - ‎‎13 ‎March ‎2021 ‏‎
Modified time - ‎‎‎14 ‎March ‎2021‎

"""

import numpy as np
import pylab as pl

# Make an array of x,y values.

x1 = [2, 3, 5, 6, 8]
x2 = [3, 4, 6, 7, 9]

y1 = [1, 5, 10, 18, 20]
y2 = [2, 6, 11, 20, 22]

# Axis limit.
pl.axis([0, 10, 0, 30]) 

# use pylab to plot x and y as red circles and blue star.
pl.plot(x1, y1,'b*', x2, y2, 'ro')

# show the plot on the screen
pl.show()
```
